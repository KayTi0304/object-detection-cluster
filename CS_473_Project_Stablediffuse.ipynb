3. Initialize the ocr text processing reader and function.

```python
# processing the ocr text processing
reader = easyocr.Reader(['en'])

def ocr_text_processing(filename, shape: list):
    #[[x cood rec1], [y cood rect1]], [[x cood rec1], [y cood rect1]]
    result = []
    # shape = [[x cood], [y cood]]
    image = Image.open(filename)
    image_cropped = image.crop((shape[0][0], shape[0][1], shape[1][0], shape[1][1]))
    result = reader.readtext(np.array(image_cropped), detail = 0)

    filtered_result = []
    for w in result:
        if w not in stop_words:
            filtered_result.append(w)

    # stem
    #stemmed_result = [ps.stem(w) for w in filtered_result]
    return filtered_result
```

4. Initialize extract function.

```python
# extracting each box set and getting the list of words
def extract(tfoutput: list):
  # tf output = [[image path][[entities][rel]...]]
    actual_result = []
    for i in tqdm(range(len(tfoutput))): # entity, # rel_attr, # rel, # many, # one, # weak entity, # ident_rel
        # for each image

        file = tfoutput[i]
        filename = file[0]

        # TODO: instead of adding to output, just make a txt file for each image

        output = [filename]
        for j in range(1, 8):
          class_name = category_labels[j-1]
          box_class_list = file[j]
          result_words_in_box_class = [] # this adds the label name to each object

          for box_class in box_class_list:

            ocr = []
            ocr.append(class_name)
            ocr.extend(ocr_text_processing(filename, box_class)) # appends list as a list itself, instead of adding it directly
            result_words_in_box_class.append(ocr)  # --> array of arrays

          if (len(result_words_in_box_class) != 0):
            # continue
            output.append(result_words_in_box_class)

        actual_result.append(output)

        '''result_file = open(filename[filename.rfind('/') + 1:filename.rfind('.')] + ".txt", "w")
        print()
        for x in range(1, len(output)):
          for erdObject in output[x]:
            result_file.write(str(erdObject) + "\n")
        result_file.close()'''

    return actual_result
```

```python
# editdistance (wip)
import editdistance
import string

# open the question text (make sure to change the path here if necessary)
f1 = open(question_path, "r", encoding="utf-8")
lines = f1.readlines()

# parse into word tokens
words_in_one_line = " ".join(lines).lower()
translator = str.maketrans("", "", ",.:()")
words_in_one_line = words_in_one_line.translate(translator)
word_tokens = words_in_one_line.split()

print("wt: ", word_tokens)

# call the edit distance
def word_substitute(target: str, max_dist: int):
  most_min_dist = float('inf')
  best_word = target
  for word in word_tokens:
      cur_dist = editdistance.eval(word, target)
      if cur_dist < most_min_dist:
        best_word = word
        most_min_dist = cur_dist

  if most_min_dist > max_dist:
    # if the best found word is too different; return original word
    return target
  else:
    return best_word

# make sure output is formatted correctly
print("output: ", output)

for box_class in output:
  box_class_list = box_class[1:] #[[...], [...], [...]], [], [], [], [], []
  for bcl in box_class_list:
    for i, a_box in enumerate(bcl):
      #print(a_box)
      for j, word in enumerate(a_box):
        terms = word.split()
        for k in range(len(terms)):
          terms[k] = terms[k].lower()
          terms[k] = word_substitute(terms[k], 2)

        a_box[j] = " ".join(terms)

      if len(a_box) == 1: # to tackle the single rel problem
        a_box.append('has')

#print("final output: ", output)
```

```python
from __future__ import division
import re
import math
from nltk.stem import PorterStemmer
import random
import numpy as np
import os

k = num_clusters # Number of clusters

global_word_set = set()  # A set of all the words that appear across all ERDs
wordlists = []  # A list of lists where the inner lists are all words in each ERD

ps = PorterStemmer()
for document in output:
    for erd_type in document:
        for object in erd_type:
            for word_index in range(1, len(object)):
                word = ps.stem(object[word_index])
                object[word_index] = word.lower()
                if ' ' in word:
                    dissect = word.split(' ')
                    object.remove(word)
                    for thing in dissect:
                        object.append(thing)
for document in output:
    for erd_type in document:
        for object in erd_type:
            for word_index in range(1, len(object)):
                word = ps.stem(object[word_index])
                object[word_index] = word.lower()
                if '_' in word:
                    dissect = word.split('_')
                    object.remove(word)
                    for thing in dissect:
                        object.append(thing)
for document in output:
    for erd_type in document:
        for object in erd_type:
            for word_index in range(1, len(object)):
                word = ps.stem(object[word_index])
                object[word_index] = word.lower()
                if '-' in word:
                    dissect = word.split('-')
                    object.remove(word)
                    for thing in dissect:
                        object.append(thing)


for document in output:
    wordlist = []
    for erd_type in document:
        for object in erd_type:
            for word_index in range(1, len(object)):
                # Stem each word and add it to this ERD's word list as well as the global word set
                word = ps.stem(object[word_index])
                wordlist.append(word)
                global_word_set.add(word)
    wordlists.append(wordlist)

# Handling term frequency
dictionaries = []  # This is the list of TF vectors for each ERD

# Iterate through each ERD's wordlist
for wordlist in wordlists:
    dictionary = {}  # The TF vector for this ERD

    # Add all words in the global list to the dictionary, initialized to 0
    for w in global_word_set:
        dictionary.update({w: 0})

    # Increments the dictionary value for each word in the ERD
    for w in wordlist:
        x = dictionary.get(w)
        x = x + 1
        dictionary.update({w: x})

    # Calculates the log term frequency the way that the project document specifies
    for w in dictionary:
        x = dictionary.get(w)
        if x != 0:
            x = math.log(x) + 1
        dictionary.update({w: x})

    dictionaries.append(dictionary)

# Format vectors to output a table
vectors = []
for dict in dictionaries:
    vectors.append(list())

for word in wordlist:
    for x in range(len(vectors)):
        vectors[x].append(dictionaries[x][word])

print(vectors)

def chooseCentroids(vectors, k):
    if k <= 0:
        return []

    centroids = []
    firstCentroid = random.randrange(0, len(vectors))
    centroids.append(firstCentroid)
    while len(centroids) < k and len(centroids) < len(vectors):
        # Choose the next centroid

        # Need to find the shortest distance from each point to the closest already chosen centroid
        distances = {}
        for curr in range(len(vectors)):
            if curr in centroids:
                continue

            # Calculate the distance from this point to all other already chosen centroids
            distsToCentroids = []
            for centroid in centroids:
                distsToCentroids.append(math.dist(vectors[centroid], vectors[curr]))

            # Find the minimum distance from this point to another already chosen centroid
            distances[curr] = min(distsToCentroids)

        # Generate random centroid proportional to the distances
        sumOfDists = sum(distances.values())
        probabilities = {}
        for x in distances.keys():
            probabilities[x] = distances[x] / sumOfDists

        rand = random.uniform(0, 1)

        newCentroid = None
        for point in probabilities.keys():
            rand -= probabilities[point]
            if rand <= 0:
                newCentroid = point
                break

        centroids.append(newCentroid)

    return centroids


def randIndex(clustering1, clustering2):
    a = 0
    b = 0
    nC2 = 0
    for i in range(len(clustering1) - 1):
        for j in range(i + 1, len(clustering1)):
            nC2 += 1
            if clustering1[i] == clustering1[j]:
                if clustering2[i] == clustering2[j]:
                    a += 1
            else:
                if clustering2[i] != clustering2[j]:
                    b += 1

    return (a + b) / nC2


def kmeans(vectors, k):
    # generate random centroids

    centroidIndexes = chooseCentroids(vectors, k)
    centroids = []
    for x in range(k):
        centroids.append(vectors[centroidIndexes[x]])

    clusters = [0] * len(vectors)

    for i in range(10): # number of iterations
        for v in range(len(vectors)):
            distances = {}
            for x in range(k):
                distances[x] = (math.dist(vectors[v], centroids[x]))

            closestCentroidDist = float('inf')
            closestCentroidIndex = -1
            for d in range(k):
                dist = distances[d]
                if dist < closestCentroidDist:
                    closestCentroidDist = dist
                    closestCentroidIndex = d

            clusters[v] = closestCentroidIndex

        for clusterIndex in range(k):
            cluster = []
            for v in range(len(clusters)):
                if clusters[v] == clusterIndex:
                    cluster.append(vectors[v])


            centroids[clusterIndex] = [float(sum(col)/len(col)) for col in zip(*cluster)]

        # print(f"cluster iteration {i+1}: {clusters}\n")

    return clusters

avg_rand = 0
total_rand = 0
num_iter = 0
for _ in range(0,1):
    actualClusters = kmeans(vectors, k)
    for i in range(k):
        nodesInCluster = list()
        for j in range(len(actualClusters)):
            if actualClusters[j] == i:
                nodesInCluster.append(str(output[j][0])[-9:-4])
                #nodesInCluster.append(sortedFileNames[j][-6:-4])
        text_output = str(nodesInCluster).replace('[', '')
        text_output = text_output.replace(']', '')
        text_output = text_output.replace('\'', '')
        print(text_output)

    correctClusters = [2, 0, 1, 1, 1, 1, 3, 0, 0, 0, 0]
    total_rand += randIndex(actualClusters, correctClusters)
    num_iter += 1
    #print('Rand index: ', randIndex(actualClusters, correctClusters))

avg_rand = total_rand / num_iter
#print(avg_rand)

```
